```python
import numpy as np
import pandas as pd
import math
import operator
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import random

random.seed(0)
# Create random matrices from normal distribution with mu = 2 and sigma = 0.01,0.1,1 respectively
mu,sigma1,sigma2,sigma3 = 2,0.01,0.1,1

# Creating 3 matrices of dimensions n × m, where n = 100 and m = 2
A1 = np.random.normal(mu, sigma1,(100,2))
A2 = np.random.normal(mu, sigma2,(100,2))
A3 = np.random.normal(mu, sigma3,(100,2))
```

Extracting each columns of three matrices

```python
    
x1=A1[:,0]
y1=A1[:,1]

x2=A2[:,0]
y2=A2[:,1]

x3=A3[:,0]
y3=A3[:,1]
```

Finding mean of the predictor variables and the target variables for calculating beta0 and beta1 values for 3 differnet datasets. 

```python

mean_x1 = np.mean(x1)
mean_y1 = np.mean(y1)

mean_x2 = np.mean(x2)
mean_y2 = np.mean(y2)

mean_x3 = np.mean(x3)
mean_y3 = np.mean(y3)

print(mean_x1, mean_x2, mean_x3)
print(mean_y1, mean_y2, mean_y3)
```

Now I calculate Beta 1(slope) and Beta 0(intercept) for the first matrix. After calculating beta 0 and beta 1, I use those values to find new predicted values. The formula used for calculating Beta 0 and Beta 1 is taken from the Machine Learning reference that is uploaded in the Learnweb.


```python
# total no.of input values
B1_1 = len(x1)

# using the formula to calculate B1 & B0 
numerator = 0
denominator = 0
for i in range(B1_1):
  numerator += (x1[i] - mean_x1) * (y1[i] - mean_y1)
  denominator += (x1[i] - mean_x1) ** 2
B1_1 = numerator / denominator #formula taken from slide 20 of Machine Learning PDF
B0_1 = mean_y1 - (B1_1 * mean_x1)

print(B1_1,B0_1)

pred_1 =[] #creating an empty list
pred1 = []
pred11 =[]
for j in range(len(x1)): 
    y1_pred = B0_1 + B1_1 * x1[j] # calculating the new predicted values using the obtained Beta0 and Beta1 values
    pred_1.append(y1_pred) # appending all the values in the list created above
    y1pred = B1_1*x1[j] #putting Beta0 as zero and finding the predicted variables
    pred1.append(y1pred) #appending all the values in the list created above
    y11pred = B0_1 + 0*x1[j]#putting Beta1 as zero and finding the predicted variables
    pred11.append(y11pred)#appending all the values in the list created above
```

Now using the same method as above I found out the values for Beta 1 and Beta 0 for the second matrix.
Then using the obtained beta values I found out the predicted values.

```python
# total no.of input values
B1_2 = len(x2)

##Implement LEARN-SIMPLE-LINREG algorithm and train it using matrix A to learn values of β0 and β1
# using the formula to calculate B1 & B0 
numerator2 = 0
denominator2 = 0
for i in range(B1_2):
  numerator2 += (x2[i] - mean_x2) * (y2[i] - mean_y2)
  denominator2 += (x2[i] - mean_x2) ** 2
B1_2 = numerator2 / denominator2
B0_2 = mean_y2 - (B1_2 * mean_x2)

print(B1_2,B0_2)

#Implement PREDICT-SIMPLE-LINREG and calculate the points for each training example in matrix A.
pred_2 =[]
pred2 = []
pred22=[]
for j in range(len(x2)): 
    y2_pred = B0_2 + B1_2 * x2[j]
    pred_2.append(y2_pred)
    y2pred = B1_2*x2[j]
    pred2.append(y2pred)
    y22pred = B0_2 + 0*x2[j]
    pred22.append(y22pred)
```

Now using the same method as above I found out the values for Beta 1 and Beta 0 for the third matrix.
Then using the obtained beta values I found out the predicted values.

```python
# total no.of input values
B1_3 = len(x3)

# using the formula to calculate B1 & B0 
numerator3 = 0
denominator3 = 0
for i in range(B1_3):
  numerator3 += (x3[i] - mean_x3) * (y3[i] - mean_y3)
  denominator3 += (x3[i] - mean_x3) ** 2
B1_3 = numerator3 / denominator3
B0_3 = mean_y3 - (B1_3 * mean_x3)

print(B1_3,B0_3)

pred_3 =[]
pred3=[]
pred33=[]
for j in range(len(x1)): 
    y3_pred = B0_3 + B1_3 * x3[j]
    pred_3.append(y3_pred)
    y3pred = B1_3*x3[j]
    pred3.append(y3pred)
    y33pred = B0_3 + 0*x3[j]
    pred33.append(y33pred)
```

### Plotting the graphs
Plotting the training and our predicted line.

```python
plt.figure(figsize=(15,10))
plt.subplot(131)
plt.title("Matrix 1 Plot")
plt.scatter(x1,y1, label='data points')
plt.plot(x1,pred_1)
plt.xlabel("Training Values")
plt.ylabel("Predicted Values")


plt.subplot(132)
plt.title("Matrix 2 Plot")
plt.scatter(x2,y2, label='data points')
plt.plot(x2,pred_2)
plt.xlabel("Training Values")
plt.ylabel("Predicted Values")

plt.subplot(133)
plt.title("Matrix 3 Plot")
plt.scatter(x3,y3, label='data points')
plt.plot(x3,pred_3)
plt.xlabel("Training Values")
plt.ylabel("Predicted Values")

plt.legend()

plt.show()
```

### Result 

From the above graphs for all the 3 datasets we can see that the data is evenly scattered and that the predicted and target values have a linear relationship. It is also observed that with increasing values of sigma the data points are more scattered.
