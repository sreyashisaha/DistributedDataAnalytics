## Matrix Multiplication

```python
import numpy as np
```

The first preliminary task is to genrate a matric with random data. Matrix 'A' has dimensions 100 * 20 i.e it has total 100 rows and 20 columns. Next we create a vector v consisting of 20 rows and one column. It is generated from the Normal distribution with mean 2 and standard deviation 0.01.

```python
# Creating a matrix A of dimensions n × m, where n = 100 and m = 20
A = np.random.random((100,20)) # Initializing the matrix with random values
print(A) #printing the matrixs created
```

```python
# Creating a numpy vector v of dimensions m × 1
mu,sigma = 2,0.01 
v = np.random.normal(mu, sigma,(20,1)) #Initializing the vector v with values from a normal distribution using μ = 2 and σ = 0.01.
print(v)
```

### Task 1
Iteratively multiplying (element-wise) each row of the matrix 'A' with vector 'v' and summing the result of each iteration in another vector c. To do this we take a variable 'row' which reads all the variable of the matrix 'A' row-wise and another varible 'col' which reads each variable of the vector v column-wise. Lastly we run a avriable named 'rows' to perfrom the multiplication between the elements and sum them in the vector 'c'.

```python
c = np.ndarray(shape = (100,1))

# iterating through rows of A
for row in range(len(A)):
   # iterating through columns of v
   for col in range(len(v[0])):
       # iterating through rows of v
       for rows in range(len(v)):
           c[row][col] += A[row][rows] * v[rows][col] # summing the result of each iteration
print(c)
```

### Task 2
Finding out the mean and standard deviation of the new vector c.


```python
mean = c.mean()
sd = c.std()
print(mean, sd)
```

### Task 3
Plotting the histogram of vector c using 5 bins.

```python
import matplotlib.pyplot as plt
plt.hist(c, bins=5,color='steelblue',edgecolor='black')
plt.title("Histogram")
```

## Grading Problem

First importing the Grades.csv file into jupyter notebook and viewing the same.

```python
import pandas as pd
grades = pd.read_csv('Grades.csv')
grades
```

There are a total of 30 students each getting scores in English, Maths, Science German and Sports.
Our task is to:
    – Compute the sum for all subjects for each student.
    – Compute the average of the point for each student. (total points are 500). 
    – Compute the standard deviation of point for each student.
    
For doing so I created a data frame and stored the file in it. Then I found out the sum and average for each student. I also found out the standard variation to measure the variation among marks obtained in 5 subjects for each student. 

```python
df = pd.DataFrame(grades)
df['Sum']=df.iloc[:,2:7].sum(axis=1)
df['Average Score']= df['Sum']/500 *100
df['Standard Deviation'] = df.iloc[:,2:7].std(axis=1)
df
```

I have then plotted the average points for all students 

```python

plt.ylabel('Average Score')
plt.plot(df['Average Score'])
plt.title("Average Points For All Students")
plt.xlabel("Student Index")
plt.show()
```
